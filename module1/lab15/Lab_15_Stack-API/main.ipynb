```python
from stackapi import StackAPI
import datetime
```

```python
SITE = StackAPI('stackoverflow')
```

#### Question 1: Find the questions and answers of last month.

```python
# your code here
x1 = datetime.datetime(2022, 2, 1)
x2 = datetime.datetime(2022, 2, 28)
questions = SITE.fetch('questions', fromdate=x1, todate=x2)
print(questions)


```

```python
answers = SITE.fetch('answers', fromdate=x1, todate=x2)
print(answers)
```

#### Question 2: Find the most voted question today with at least a score of 5 and tagged with 'python'. 

```python
# your code here
question2 = SITE.fetch('questions', fromdate=datetime.datetime(2022, 3, 6), todate=datetime.datetime.now(),
                      tagged='python', min=5, sort="votes")
print(question2)
```

#### Question 3: Find the answers with id 6784 and 6473.

```python
# your code here
badges = SITE.fetch('badges', ids=[6784, 6473])
print(badges)
```
